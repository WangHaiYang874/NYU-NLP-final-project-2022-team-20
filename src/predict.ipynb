# Evaluation

#### remark:
The personality classification of mbti type can be boils down to 4 binary classifications
- introverts versus extroverts
- intuition versus sensing
- thinking versus feeling
- judge versus perceive

Our model achieves $\bold{\ge 80\%}$ accuracy on the classification of the later three binary classification. And it achieves $\bold{\ge 80\%}$ accuracy for introverts versus extroverts. 

We have the following observations:
- the accuracy of `I versus E` can be increased by 
  - using a larger dictionary and therefore more tfidf dimensions
  - bigger training sample: the accuracy was 74% with training sample of 2000, the accuracy was >80% with training sample of 200000
- the accuracy of three other binary classifications will become 100% when the training sample is merely 2000. 

This reveals the true hard task of personality prediction is perhaps trying to distinguish introverts and extroverts. 

We do not have a satisfiable explanation of why our model performed so well.

```python
train_X = pred.feature_extractor.get_features(traindf.body)
train_y = traindf.mbti_type
```

```python
cog_funs23 = {k:v for k,v in pred.first_layer.items() if sum(['_' == i for i in k]) in [2,3]}
```

```python
def get_cog_funs(train_X):
    return np.array([cog_funs23[k].predict_proba(train_X)[:,0] for k in sorted(list(cog_funs23))]).T
```

```python
train_X2 = get_cog_funs(train_X)
```

```python
clf = RandomForestClassifier(n_estimators=5000,criterion='gini')
clf.fit(train_X2, train_y)
```

```python
test_X2 = get_cog_funs(test_X)
```
